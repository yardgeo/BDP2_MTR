The following GraphQL string fetches information about some proteins:

```
{
  entries(entry_ids: ["3dnb", "3lpv", "1ckt"]) {
    rcsb_id
    rcsb_primary_citation {
      pdbx_database_id_DOI
      pdbx_database_id_PubMed
    }
  }
}
```

Write and then run code that does the following:

1. Use the string above to query the PDB GraphQL API, available at http://data.rcsb.org/graphql to fetch data about the polymer entities. For each protein, get the DOI and the PubMed ID. 

2. Query PubMed to get the primary citation for eath protein, using the endpoint https://api.ncbi.nlm.nih.gov/lit/ctxp/v1/pubmed/?format=citation&id=ID, replacing `ID` with the PubMed ID. You should get the citation in the American Medical Association (AMA) style (look at the `ama` key in the returned output from PubMed and then select the `orig` format).

   Print a summary of the information you obtained, formatting it like this:

   `Citation for <protein_id> (DOI: <DOI>): <citation>`

3. (optional) For each citation retrieve and print the text of the abstract, querying the endpoint  https://www.ncbi.nlm.nih.gov/research/bionlp/RESTful/pubmed.cgi/BioC_json/ID/unicode, where `ID` is the PubMed ID. You will have to explore the returned data structure to find and extract the _abstract_.

````python
import requests
````

````python
query = '''
{
  entries(entry_ids: ["3dnb", "3lpv", "1ckt"]) {
    rcsb_id
    rcsb_primary_citation {
      pdbx_database_id_DOI
      pdbx_database_id_PubMed
    }
  }
}
'''

r = requests.get('https://data.rcsb.org/graphql?query=%s' % requests.utils.requote_uri(query))
j = r.json()    #with the json format we get a dictionary
#print(j)       #i used this print to see how the informations are organized
````

````python
DOIs = []
pm_IDs = []
prot_list = ["3dnb", "3lpv", "1ckt"]

for i in range(len(prot_list)):    #the entries dictionary has a list as a key, and each element is one of the proteins, but they are ordered as requested, so i used an index to call them
    DOI = j["data"]["entries"][i]["rcsb_primary_citation"]["pdbx_database_id_DOI"]
    pm_ID = j["data"]["entries"][i]["rcsb_primary_citation"]["pdbx_database_id_PubMed"]
    
    DOIs.append(DOI)     #each DOI and each pm_IDs is appendend to the lists
    pm_IDs.append(pm_ID)

print("DOIs: ",DOIs)
print("PubMed_IDs: ",pm_IDs)
````

````python
#for each one of the protein we will make a REST call to PubMed
for i,protein in enumerate(prot_list):
    r_pubmed = requests.get("https://api.ncbi.nlm.nih.gov/lit/ctxp/v1/pubmed/?format=citation&id=%s" % pm_IDs[i]) #we used the pubmed ids stored in the list, again using the indexes
    data = r_pubmed.json()   #each result is stored in a data dictionary
    print("Citation for <%s> (DOI: <%s>): <%s>\n" % (protein,DOIs[i],data["ama"]["orig"]))    #here we get the citation with the ama formatting
````

````python
#for 8.3
for i,protein in enumerate(prot_list):
    response = requests.get("https://www.ncbi.nlm.nih.gov/research/bionlp/RESTful/pubmed.cgi/BioC_json/{ID}/unicode".format(ID=pm_IDs[i]))
    json_r = response.json()
    for passage in json_r['documents'][0]['passages']:
        if passage['infons']['type'] == 'abstract':
            print("Abstract for ", protein)
            print(passage["text"])
````
